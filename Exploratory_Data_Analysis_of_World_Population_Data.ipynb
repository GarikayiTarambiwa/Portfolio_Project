```python
#Country with the Lowest population per year

numeric_cols = [str(year) for year in range(1960, 2023)]
results = []

for col in numeric_cols:
    col_data = df[col].dropna()
    min_val = col_data.min()
    country = col_data.idxmin()
    results.append({"Country": country, "Year": col, "Population": min_val})

# Create a DataFrame from the results and display it.
pd.DataFrame(results)
```

```python
numeric_cols = [str(year) for year in range(1960, 2023)]
results = []

for col in numeric_cols:
    col_data = df[col].dropna()
    max_val = col_data.max()
    country = col_data.idxmax()
    results.append({"Year": col, "Country": country, "Population": max_val})

pd.DataFrame(results)
```

```python

df_years = df.loc[:, '1960':'2022']

df_years.T.plot(kind='line', legend=False)  

plt.title('Total Population from 1960 to 2022')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.show()

```

```python


df_years = df.loc[:, '1960':'2022']

# Calculate the annual growth rate
df_growth = df_years.pct_change(axis='columns') * 100  # pct_change calculates the percentage change

# The growth rate will be NaN for the first year (1960) since there's no previous year to compare it to,
# so you might want to drop that column
df_growth = df_growth.drop(columns='1960')
```

```python
df_growth
```

```python

countries_to_plot = ['India', 'China']

df_growth.loc[countries_to_plot].T.plot(kind='line')

plt.title('Annual Population Growth Rate from 1961 to 2022')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend(title='Country Name')
plt.grid(True)
plt.show()
# Observe that the growth rate in China exhibited a sharp increase between 1961 to 1971
# This surge was followed by a substantial decline in growth rate from 1971 until mid-1970s.
# Post 1985, the Chinese growth rate has been on a slow and steady decline till 2022.
# On the other hand, India's growth rate has seen a moderate, consistent decrease from 1961 through to 2022.
# Despite the decrease, India's growth rate decline is not as significant as China's.
# As a result, in 2022, India's population overtook China's population.
```
