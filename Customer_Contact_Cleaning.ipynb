```python
import numpy as np
import pandas as pd
import re
```

```python

df = pd.read_excel(r"C:\Users\garik\OneDrive\Documents\data analysis\Data Cleaning\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python

def remove_non_alphanumeric(s):
    if pd.notnull(s) and isinstance(s, str):
        return re.sub('[^a-zA-Z0-9]', '', s)
    else:
        return s

df.loc[:, 'Last_Name'] = df['Last_Name'].apply(remove_non_alphanumeric)
df
```

```python
def format_phone_number(x):
    x = str(x)
    y = re.sub('[^a-zA-Z0-9]', '', x)
    z = y[:3]+'-'+y[3:6]+'-'+y[6:]
    return z

df.loc[:, 'Phone_Number'] = df["Phone_Number"].apply(format_phone_number)
df["Phone_Number"]
```

```python
df.loc[:, 'Paying Customer'] = df["Paying Customer"].replace(to_replace = "Yes", value = "Y")
df.loc[:, 'Paying Customer'] = df["Paying Customer"].replace(to_replace = "No", value = "N")
df["Paying Customer"]
```

```python
df.loc[:, 'Do_Not_Contact']= df["Do_Not_Contact"].replace(to_replace = "Yes", value = "Y")
df.loc[:, 'Do_Not_Contact'] = df["Do_Not_Contact"].replace(to_replace = "No", value = "N")
df["Do_Not_Contact"]
```
